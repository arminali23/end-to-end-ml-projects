```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
X = np.random.rand(100,1)
y = 3 * X.squeeze() + 7 + np.random.randn(100) * 0.8

plt.scatter(X,y, alpha=0.6, color='steelblue')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Dataset')
plt.show()
```

```python
w = np.random.randn()
b = np.random.randn()

print(f"initial w : {w:.4f}, b: {b:.4f} ")
```

```python
def predict (X, w, b):
    return X.squeeze() * w + b
```

```python
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

```python
def compute_gradients(X, y, w, b):
    n = len(y)
    X = X.squeeze()
    y_pred = X * w + b
    error = y_pred - y
    
    dw = (2/n) * np.dot(X, error)
    db = (2/n) * np.sum(error)
    
    return db,dw
```

```python
learning_rate = 0.1
epochs = 100
loss_history = []

for epoch in range(epochs):
    y_pred = predict(X, w, b)
    loss = mse_loss(y, y_pred)
    loss_history.append(loss)

    dw, db = compute_gradients(X, y, w, b)

    w = w - learning_rate * dw
    b = b - learning_rate * db

    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1}/{epochs} | Loss: {loss:.4f} | w: {w:.4f} | b: {b:.4f}")
```

```python
plt.plot(loss_history, color = 'crimson')
plt.xlabel('epoch')
plt.ylabel('mse loss')
plt.title('loss curve')
plt.grid(True)
plt.show()
```
